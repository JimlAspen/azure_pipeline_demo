```python
from azure.storage.blob import BlobServiceClient
from azureml.core import Workspace

ws = Workspace.from_config()
# Get the default Key Vault associated with your workspace
kv = ws.get_default_keyvault()

connection_string = kv.get_secret(name="blob-conn")

container_name = "raw"
local_file_path = "../azure_crash_course/breast_cancer.csv"
blob_name = "breast_cancer.csv"

# Initialize using the explicit connection string
blob_service_client = BlobServiceClient.from_connection_string(connection_string)
blob_client = blob_service_client.get_blob_client(container=container_name, blob=blob_name)

print("Uploading file via Connection String...")
with open(local_file_path, "rb") as data:
    blob_client.upload_blob(data, overwrite=True)
print("Upload complete!")
```
